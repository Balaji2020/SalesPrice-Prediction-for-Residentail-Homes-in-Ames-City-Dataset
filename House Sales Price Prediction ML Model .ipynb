### 3.3 Dealing with Missing Values

```python
house_data['PoolQC']=house_data['PoolQC'].fillna('None')               #Replace NA Values in 'PoolQC' Data field with None Values for No Pool
house_data['MiscFeature']=house_data['MiscFeature'].fillna('None')     #Replace NA Values in 'MiscFeature' Data field with None Values for No MiscFeature
house_data['Alley']=house_data['Alley'].fillna('None')                 #Replace NA Values in 'Alley' Data field with None Values for No Alley
house_data['Fence']=house_data['Fence'].fillna('None')                 #Replace NA Values in 'Fence' Data field with None Values for No Fence
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    house_data[col]=house_data[col].fillna('None')                     #Replace NA Values in 'GarageType', 'GarageFinish', 'GarageQual' and 'GarageCond' data field with None value
   
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):                #Replace NA Values in 'GarageYrBlt', 'GarageArea' and 'GarageCars' data field with 0 values
    house_data[col] = house_data[col].fillna(0)

for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    house_data[col] = house_data[col].fillna(0)                        #Replace NA Values in 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath' and 'BsmtHalfBath' data field with 0 Values
    
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    house_data[col]=house_data[col].fillna('None')                     #Replace NA Values in 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1' and 'BsmtFinType2' data field with None Values
    
house_data['MasVnrType'] = house_data['MasVnrType'].fillna('None')     #Replace NA Values in 'MasVnrType' with None Values.
house_data['MasVnrArea'] = house_data['MasVnrArea'].fillna(0)          #Replace NA Values in 'MasVnrArea' with 0 Values.
house_data['Electrical'] = house_data['Electrical'].fillna(house_data['Electrical'].mode()[0])  #Replace NA Values in 'Electrical' data field with mode of values
house_data['FireplaceQu'] = house_data['FireplaceQu'].fillna('None')   #Replace NA Values in 'FireplaceQu' data field with None Values as NA means No Fireplace
```

__As the the area of each street connected to the house property most likely have a similar area to other houses in its neighborhood , we can fill in missing values by the median LotFrontage of the neighborhood__.

```python
house_data['LotFrontage']=house_data.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

__Introducing new feature TotalSF(total square footage) to determine house prices with the help of total area of basement, first and second floor areas of each house__.

```python
house_data['TotalSF'] = house_data['TotalBsmtSF'] + house_data['1stFlrSF'] + house_data['2ndFlrSF']
```

```python
feature_cols = ['TotalSF','GrLivArea','GarageArea']            # create a Python list of feature names
X = house_data1[feature_cols]                     # use the list to select a subset of the original DataFrame-+
```

```python
y = house_data1.SalePrice
```

- __Splitting X and y into training and test datasets__. <br/>

```python
from sklearn.cross_validation import train_test_split

def split(X,y):
    return train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
X_train, X_test, y_train, y_test=split(X,y)
print('Train cases as below')
print('X_train shape: ',X_train.shape)
print('y_train shape: ',y_train.shape)
print('\nTest cases as below')
print('X_test shape: ',X_test.shape)
print('y_test shape: ',y_test.shape)
```

- __Linear Regression without Gridsearch CV__. <br/>

```python
def linear_reg( X, y, gridsearch = False):
    
    X_train, X_test, y_train, y_test = split(X,y)
    
    from sklearn.linear_model import LinearRegression
    linreg = LinearRegression()
    
    if not(gridsearch):
        linreg.fit(X_train, y_train) 

    else:
        from sklearn.model_selection import GridSearchCV
        parameters = {'normalize':[True,False], 'copy_X':[True, False]}
        linreg = GridSearchCV(linreg,parameters, cv = 10)
        linreg.fit(X_train, y_train)                                                           # fit the model to the training data (learn the coefficients)
        print("Mean cross-validated score of the best_estimator : ", linreg.best_score_)  
        
        y_pred_test = linreg.predict(X_test)                                                   # make predictions on the testing set

        RMSE_test = (metrics.mean_squared_error(y_test, y_pred_test))                          # compute the RMSE of our predictions
        print('RMSE for the test set is {}'.format(RMSE_test))
        R2_Grid_CV_test  = metrics.r2_score(y_test, y_pred_test)
        print('R2 score for the test set is {}'.format(R2_Grid_CV_test))

    return linreg

X = house_data1[feature_cols]  
y = house_data1.SalePrice
linreg = linear_reg(X,y)
```

- __Interpreting Model Coefficients__. 

```python
feature_cols.insert(0,'Intercept')
print(feature_cols)
coef = linreg.coef_.tolist()
print(coef)
coef.insert(0, linreg.intercept_)
print(coef)
```

```python
eq1 = zip(feature_cols, coef)

for c1,c2 in eq1:
    print(c1,c2)
```

### Observation:

- We can see that all the 3 input feature have a __positive__ coefficient value that's means if there is an increment in one of the feature leads an increament in the __SalePrice__ of the house.

- the final linear function equation for determining the __SalePrice__ of the house in the given dataset is as follows.

   __SalePrice(y)__ = -0.00329 + 0.6130 `*` __TotalSF__ +  0.1042 `*` __GrLivArea__ + 0.2328 `*` __GarageArea__. 


- __SalePrice Prediction for train and test dataset Using LR Model__

```python
y_pred_train = linreg.predict(X_train)     # make predictions on the training set
y_pred_test = linreg.predict(X_test)       # make predictions on the testing set
```
